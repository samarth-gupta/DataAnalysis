```python
#Question
#Which team has won/lost maximum number of matches while chasing?
```

```python
chase_win = {}
chase_lost = {}
```

```python
for index,row in matches.iterrows():
    team1 = row['team1']
    team2 = row['team2']
    winner = row['winner']
    result = row['result']
    if result == 'tie':
        continue
    if team2 == winner:
        if chase_win.get(team2) == None:
            chase_win[team2] = 1
        else:
            chase_win[team2] += 1
    elif team1 == winner:
        if chase_lost.get(team2) == None:
            chase_lost[team2] = 1
        else:
            chase_lost[team2] += 1
```

```python
chase_win
```

```python
max(chase_win, key=chase_win.get)
```

```python
max(chase_lost, key=chase_lost.get)
```

```python
#Answer: Kolkata Knight Riders ( Winner ), Delhi Daredevils ( Lost )
```

```python
#Question
#Which toss decision is best for each ground?
```

```python
toss_win = {}
```

```python
for index,row in matches.iterrows():
    team1 = row['team1']
    team2 = row['team2']
    winner = row['winner']
    result = row['result']
    venue = row['venue']
    if result != 'normal':
        continue
    if team1 == winner:
        if toss_win.get(venue) == None:
            toss_win[venue] = 1
        else:
            toss_win[venue] += 1
    elif team2 == winner:
        if toss_win.get(venue) == None:
            toss_win[venue] = -1
        else:
            toss_win[venue] -= 1
        
```

```python
toss_win
```

```python
stats = []
```

```python
for key,value in toss_win.items():
    stats.append( ( value,key, ('bat first' if value > 0 else ( 'bowl first' if value < 0 else 'equally beneficial' )) ))
```

```python
stats.sort()
```

```python
print( format('Net Value','<7'), '\t', format('Stadium/Venue','<50'), '\t', format('Decision','<10') )
for val,stad,decision in stats:
    print( format(val,'<7'), '\t', format(stad,'<50'), '\t', format(decision,'<10') )
```

```python
#Answer ( as per majority of wins/losses )
# In stats, if net value of venue is negative then Bowling first is better, if positive then Batting first.
# If net value is zero, then ( so far ) it is equally better to Bat or Bowl first.
```

```python
#Question
#Which player has won maximum number of MoM awards while chasing?
```

```python
awards_chase = {}
```

```python
for index,row in matches.iterrows():
    match_id = row['id']
    team1 = row['team1']
    team2 = row['team2']
    winner = row['winner']
    result = row['result']
    player = row['player_of_match']
    team = ''
    match_deliveries = deliveries.loc[ deliveries['match_id'] == match_id ]
    
    for index,row in match_deliveries.iterrows():
        
        batting_team = row['batting_team']
        bowling_team = row['bowling_team']
        batsman = row['batsman']
        non_striker = row['non_striker']
        bowler = row['bowler']
        
        if player == batsman or player == non_striker:
            team = batting_team
            break
        elif player == bowler:
            team = bowling_team
            break
    
    if team == team2:
        if awards_chase.get(player) == None:
            awards_chase[player] = 1
        else:
            awards_chase[player] += 1
            
    
```

```python
#Question
#Which bowler has hit for maximum number of 1s/2s/3s/4s/6s?
```

```python
hits = {}
```

```python
for index,row in deliveries.iterrows():
    runs = str(row['batsman_runs'])
    bowler = row['bowler']
    if hits.get(runs) == None:
        hits[runs] = {}
    if hits[runs].get(bowler) == None:
        hits[runs][bowler] = 1
    else:
        hits[runs][bowler] += 1
```

```python
for runs in ['0','1','2','3','4','5','6']:
    print("Maximum number of Runs =",runs,"given by =", max(hits[runs], key=hits[runs].get))
```

```python
#Answer Shown above, players who gave max number of specified runs.
```

```python
#Question
#Which batsman got stumped out maximum number of times?
```

```python
stumped_cnt = {}
```

```python
#Extra Question 3
#Which player has played for maximum number of different teams ( in total )
```

```python
player_teams = {}
```

```python
for index,row in deliveries.iterrows():
    
    batting_team = row['batting_team']
    bowling_team = row['bowling_team']
    batsman = row['batsman']
    non_striker = row['non_striker']
    bowler = row['bowler']
    
    if player_teams.get(batsman) == None:
        player_teams[batsman] = set()
    if player_teams.get(non_striker) == None:
        player_teams[non_striker] = set()
    if player_teams.get(bowler) == None:
        player_teams[bowler] = set()
    
    player_teams[batsman].add(batting_team)
    player_teams[non_striker].add(batting_team)
    player_teams[bowler].add(bowling_team)        
```

```python
max(player_teams,key=( lambda x : len(player_teams.get(x) )) )
```

```python
#Answer AJ Finch has played for maximum number of distinct teams.
```

```python
#Extra Question 5
#Calculate average runs scored per match by each team
```

```python
team_runs_matches = {}
```

```python
for index,row in matches.iterrows():
    match_id = row['id']
    team1 = row['team1']
    team2 = row['team2']
    match_deliveries = deliveries.loc[ deliveries['match_id'] == match_id ]
    run1 = 0
    run2 = 0
    
    for index,row in match_deliveries.iterrows():
        total_runs = row['total_runs']
        batting_team = row['batting_team']
        if batting_team == team1:
            run1 += total_runs
        else:
            run2 += total_runs
    
    if team_runs_matches.get(team1) == None:
        team_runs_matches[team1] = []
    if team_runs_matches.get(team2) == None:
        team_runs_matches[team2] = []
        
    team_runs_matches[team1].append(run1)
    team_runs_matches[team2].append(run2)
```

```python
print( format('Team Name',"<30"), format('Average Runs',"<6"), '\t', 'Number of Matches' )
for teams,val in team_runs_matches.items():
    avg = sum(val)/len(val)
    print( format(teams, "<30" ), format(avg,"<6f"), "\t", len(val))
    
```

```python
#Answer The average runs per match scored by each team are shown above ( along with number of matches played )
```
